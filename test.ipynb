```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('data/reidin_merged.csv')
df['Date'] = pd.to_datetime(df["Date"], format="%d-%m-%Y", exact=False).dt.date
df.head()
```

```python
df['Location'].isna().sum()
```

```python
df[['Sub', 'Area']] = df['Location'].str.split(', ', expand=True)
```

```python
df.head()
```

```python
df['Area'].isna().sum()
```

```python
df['Area'].fillna('', inplace=True)
df['Area'].value_counts()
```

```python
import requests

url = "https://viewit.ae/api/locations"
resp = requests.get(url).json()
resp
```

```python
locs = []
for i in resp['Data']:
    locs.append(i['location_name'])
```

```python
locs = list(dict.fromkeys(locs))
sorted(locs)
```

```python
locs = sorted(locs)[1:]
locs
```

```python
# LIST OF LOCATIONS NOT FOUND IN VIEWIT DATABASE BUT EXISTS IN REIDIN DATA
for i in df['Area'].unique():
    if i not in locs:
        print(i)
```

```python
# LIST OF LOCATIONS NOT FOUND IN REIDIN DATA BUT EXISTS IN VIEWIT DATA
for i in locs:
    if i not in df['Area'].unique():
        print(i)
```

```python
added = []
notfound = []

for location in df['Area'].unique():
    if location in locs:
        df.loc[df['Area'] == location, 'Link'] = f"https://viewit.ae/listings?purpose=Buy&page=1&pageSize=24&location={ '%20'.join(location.split(' ')) }"
        added.append(location)
    else:
        notfound.append(location)

for i in added:
    print(f" Added link for {i}")
print('')
for j in notfound:
    print(f"Area {j} not found in the list of locations.")
```

```python
def add_link(location: str, link: str) -> None:
    df.loc[df['Area'] == location, 'Link'] = f"https://viewit.ae/listings?purpose=Buy&page=1&pageSize=24&location={ link if link else '%20'.join(location.split(' ')) }"
    print('Added Link for', location)
    return df[df['Area'] == location].head(2)
```

```python
def linkify(viewit_area_name: str) -> str:
    return '%20'.join(viewit_area_name.split())
```
